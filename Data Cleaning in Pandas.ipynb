```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Taozend\repos\ata\Datasets\Customer Call List.xlsx")
df
```

First, we're going to drop the duplicate entry found at the end of the dataset. We will be overwriting the file 'df' in this tutorial, but it is best practice to KEEP THE ORIGINAL dataset. This will allow you to make comparisons while cleaning. The originals and the cleaned files should probably also be kept in separate folders.

```python
df = df.drop_duplicates()
df
```

We should get rid of any columns that do not provide any useful information for the purpose of the calls. Not_Useful_Column is a good candidate for this.

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

The ID looks good. The first name looks good. The last name has a few inconsistencies that should be addressed. Since we are dealing only with that column, we will specify it.

We will use str.strip() to clean it up. Variations include str.lstrip() and str.rstrip() which will target specifically the left side or the right side, respectively. str.strip() will include both sides. It does take an argument that identifies a regex series of characters to be removed.

df['Last_Name'].str.lstrip('/')
df['Last_Name'].str.rstrip('_')

By the way, we will not be saving the following change with the method we used before because we would end up ONLY with that column. Very bad.

```python
df['Last_Name'] = df['Last_Name'].str.strip('123._/') # This will strip numbers, periods, underscores, and forward slashes.
df
```

Now, we'll target the phone numbers. We'll start by stripping all the values down to numbers only, and then we will format it.

As an argument for replace(), we are using regex again to say we want to replace any character that IS NOT a lowercase letter, an uppercase letter, or a number and replace them with nothing. In other words, it will just be removed. The caret (^) indicates "not".

Note: By default, the regex argument for replace has recently changed from True to False. If you want to use regex in your argument, be sure to set regex to True, otherwise it will just treat the pattern as a literal string.

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex=True)

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
# The code above was run to convert the column to a string so that the phone numbers could be formatted with dashes.

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
# This is taking each string of numbers and inserting dashes. For cells that contained "N/a" and "NaN", it just added two
# dashes to the end, so we'll remove those below.

df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

Next, we have the Address column. Let's assume it needs to be split into separate columns for the street, the state/territory/planet, and the zip code (if applicable). Fortunately, these are separated by commas, so we will specify that as the delimiter. Then, we will name the columns. The original address column, by the way, is preserved here.

```python
df[['Street_Address', 'More_General_Location', 'Zip_Code']] = df['Address'].str.split(',',expand=True)
df
```

The Paying Customer column should all be "Yes" or "No". Notice that we are first changing any "Yes" and "No" to "Y" and "N". This is because if we were to attempt to change any "Y" to a "Yes", for example, it would look for any "Y" and change that character into "Yes", and you would end up with "Yeses" wherever there was a "Yes". So, we will get the applicable cells down to a single character before changing it to a word response.

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Paying Customer'] = df['Paying Customer'].str.replace('Y', 'Yes')
df['Paying Customer'] = df['Paying Customer'].str.replace('N', 'No')
df
```

We will do the exact same thing to "Do_Not_Contact" column.

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Y', 'Yes')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('N', 'No')
df
```

That takes care of cleaning up the columns individually. Now, we will address the dataset as a whole. We'll say we were asked to remove any NaN, "N/a", etc. and replace them with nothing. Our method above gave us a "No/a" entries, as well, so we'll have to remember to take care of that.

Because NaN is a type all its own and not a string, we have to handle that differently. For NaN, we will use the fillna function. For everything else, we will still use the replace function.

We do not need ".str" as part of replace because we are dealing with the dataset as a whole.

```python
df = df.fillna('')

df = df.replace('N/a','')
df = df.replace('No/a','')
df
```

We'll say that our call center wants a list only of numbers that can be contacted, so we'll need to remove any records that are Do Not Call and any that don't have a number to call.

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Yes':
        df.drop(x, inplace=True)

df
```

```python
# That removed those on the Do Not Call list. Now, we will remove any records without a phone number.

# df_new = df.dropna(subset=['Phone_Number'], inplace=True, ignore_index=True)
# The code above was an attempt to remove empty strings from the Phone_Number column, but it removed every record.

df = df[df['Phone_Number'] != '']
```

Finally, before we hand it off, let's reset the index for the call center. We will use drop=True as an argument so it doesn't just create a new column with the current index.

```python
df.reset_index(drop=True)
```
